```python
some_data = np.random.normal(0, 2, size=500)
```

```python
sorted_some_data = np.sort(some_data)
np.max(sorted_some_data)
```

```python
#some_data = np.sort(some_data)
plt.xlabel('independent variable')
plt.ylabel('dependent variable')
plt.subplot(2, 1, 1)
noise = np.random.normal(0, 1, 500 )
noisy_data = sorted_some_data+noise
plt.plot(noisy_data, label='original')
plt.legend()
plt.subplot(2, 1, 2)
plt.xlabel('independent variable')
plt.ylabel('dependent variable')
plt.plot((noisy_data - np.min(noisy_data))/(np.max(noisy_data) - np.min(noisy_data)), label='normalised')
plt.legend()
```
